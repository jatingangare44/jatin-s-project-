# **Univariate Analysis**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Extract the height data
height_data = df['Height_(cm)']

# Descriptive statistics
mean = height_data.mean()
median = height_data.median()
mode = height_data.mode()[0]
std_dev = height_data.std()

print(f"Mean Height: {mean}")
print(f"Median Height: {median}")
print(f"Mode Height: {mode}")
print(f"Standard Deviation Height: {std_dev}")

# Generate a count plot for height
plt.figure(figsize=(10, 6))
sns.countplot(x=df['Height_(cm)'], color='skyblue')
plt.title("Height Count Plot")
plt.xlabel("Height_(cm)")
plt.ylabel("Count")
plt.show()

# Generate a histogram for height
plt.figure(figsize=(10, 6))
plt.hist(height_data, bins=20, color='skyblue', edgecolor='black')  # You can adjust the number of bins as needed
plt.title("Height Distribution")
plt.xlabel("Height_(cm)")
plt.ylabel("Frequency")
plt.show()
```

# **Bivariate Analysis**

```python

male_data = df[df['Sex'] == 'Male']
female_data = df[df['Sex'] == 'Female']
plt.figure(figsize=(10, 6))

plt.scatter(male_data['Height_(cm)'], male_data['Weight_(kg)'], color='blue', alpha=0.5, label='Male')

plt.scatter(female_data['Height_(cm)'], female_data['Weight_(kg)'], color='red', alpha=0.5, label='Female')

plt.title("Scatter Plot: Height_(cm) vs. Weight_(kg) by Gender")
plt.xlabel("Height_(cm)")
plt.ylabel("Weight_(kg)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assign numerical values to the categories in General_Health
# You can customize this mapping based on your data if needed
health_mapping = {
    'Poor': 1,
    'Fair': 2,
    'Good': 3,
    'Very Good': 4,
    'Excellent': 5
}

# Create a new column in the DataFrame with the numerical values for General_Health
df['General_Health_Numeric'] = df['General_Health'].map(health_mapping)

# Create a scatter plot with General_Health on the x-axis and BMI on the y-axis
plt.figure(figsize=(10, 6))
sns.scatterplot(x='General_Health_Numeric', y='BMI', data=df, palette='viridis')
plt.title("Scatter Plot: General_Health vs. BMI")
plt.xlabel("General_Health")
plt.ylabel("BMI")
plt.xticks(ticks=[1, 2, 3, 4, 5], labels=health_mapping.keys())  # Set custom x-axis labels
plt.grid(True)
plt.show()

# Remove the added numerical column to avoid interference with the original data
df.drop('General_Health_Numeric', axis=1, inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a bar plot with hue (Gender)
plt.figure(figsize=(10, 6))
sns.barplot(x='General_Health', y='Weight_(kg)', hue='Sex', data=df, ci=None, palette='Set2')
plt.title("Distribution of Health and Weight")
plt.xlabel("General_Health")
plt.ylabel("Weight_(kg)")
plt.show()
```

```python
df.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select the continuous variables for correlation analysis
continuous_columns = [
    'Height_(cm)',
    'Weight_(kg)',
    'BMI'
]

# Calculate the correlation matrix for selected columns
corr_matrix = df[continuous_columns].corr()

# Generate a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Separate DataFrame based on General_Health
poor_health_df = df[df['General_Health'] == 'Poor']
fair_health_df = df[df['General_Health'] == 'Fair']
good_health_df = df[df['General_Health'] == 'Good']
very_good_health_df = df[df['General_Health'] == 'Very Good']
excellent_health_df = df[df['General_Health'] == 'Excellent']

# Display the first few rows of each health category
print("Poor Health:")
print(poor_health_df.head())

print("\nFair Health:")
print(fair_health_df.head())

print("\nGood Health:")
print(good_health_df.head())

print("\nVery Good Health:")
print(very_good_health_df.head())

print("\nExcellent Health:")
print(excellent_health_df.head())
```

```python
import matplotlib.pyplot as plt

# Calculate the distribution of 'General_Health' categories
health_distribution = df['General_Health'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(health_distribution, labels=health_distribution.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribution of General Health")
plt.show()
```

# Multivariate Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select the variables you want to include in the pair plot
selected_columns = [
    'General_Health',
    'Height_(cm)',
    'Weight_(kg)',
    'BMI',
    'Exercise',
    'Age_Category'
]

# Create a pair plot
sns.pairplot(df[selected_columns], diag_kind='kde')
plt.suptitle("Pair Plot of Selected Variables", y=1.02)
plt.show()
```
